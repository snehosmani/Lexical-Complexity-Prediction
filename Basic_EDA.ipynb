```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('lcp_single_train.tsv.txt',sep='\t')
data_test=pd.read_csv('lcp_single_test.tsv.txt',sep='\t')
```

```python
counts = data['corpus'].value_counts()
counts_test = data_test['corpus'].value_counts()
```

```python
fig, ax = plt.subplots()
ax.bar(counts.index, counts.values)

# Set the chart title and axis labels
ax.set_title('Count of Samples by Corpus- training dataset')
ax.set_xlabel('Corpus')
ax.set_ylabel('Count')

# Display the chart
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(counts_test.index, counts_test.values)

# Set the chart title and axis labels
ax.set_title('Count of Samples by Corpus- training dataset')
ax.set_xlabel('Corpus')
ax.set_ylabel('Count')

# Display the chart
plt.show()
```
